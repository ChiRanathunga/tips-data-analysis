# Tips Data Analysis Project

## Objective
Analyze customer bills and tipping behavior.

## Tools Used
- Pandas
- Matplotlib

## Conclusion

This analysis shows that tipping behavior is strongly related to the total bill amount. Customers tend to give higher tips when their total bill is higher. Additionally, there are noticeable differences in spending patterns across different days of the week, with weekends (especially Sunday and Saturday) having higher average bills. Gender-based differences in tipping are present but relatively small. Overall, the dataset reveals clear patterns in customer spending and tipping behavior.


```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()
df.info()
df.describe()
```

```python
df['total_bill'].mean()
```

```python
df['tip'].mean()
```

```python
df['sex'].value_counts()
```

```python
df['day'].value_counts(ascending = False)
```

```python
import matplotlib.pyplot as plt

df['total_bill'].plot(kind='hist')
plt.title("Total Bill Distribution")
plt.show()
```

```python
df.plot(kind='scatter', x='total_bill',y='tip' )
plt.title('Total Bill vs Tip')
plt.show()
```

#insights
1. Most customers give tips when the total bill is between 10 and 20.
2. As the total bill increases, the tip amount also increases.
3. The average tip is between 4 and 5.


```python
df.groupby(['day']).mean('total_bill').sort_values(by='total_bill',ascending=False)
```

```python
df.groupby(['sex']).mean('tip').sort_values(by='tip',ascending=False),['tip','sex']
```

```python
df.plot(kind='scatter',x= 'smoker',y='tip')
plt.title('smoker vs tip')
plt.show()
```

1. Which day has the highest average bill?
 - Sunday
2. Do males or females tip more?
 - Males
3. Does smoking affect tip amount?
 - yes, smokers relatively give less tip than non smokers.

```python
df.groupby('day')['total_bill'].mean().sort_values(ascending=False)
```

```python
df.groupby('sex')['tip'].mean().sort_values(ascending=False)
```

```python
df.groupby('day')['total_bill'].mean().plot(kind='bar')
plt.title('Average Bill By Day')
plt.show()
```

```python
df.groupby('sex')['tip'].mean().plot(kind='bar')
plt.title('Average Tip By Gender')
plt.show()
```

#insights


1. Sunday has the highest average bill.
2. Friday has the lowest average bill.
3. The average tip given by males is slightly higher than that given by females.
4. Saturday has a higher average bill than Thursday.
